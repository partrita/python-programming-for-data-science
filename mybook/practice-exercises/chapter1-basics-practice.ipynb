### 10.

문자열이 포함된 `언어` 변수가 주어지면 `if/elif/else`를 사용하여 다음과 같은 프로그램을 작성하세요.
- "나는 뱀을 사랑해요!"를 반환합니다. `언어`가 `"python"`인 경우(모든 종류의 대문자)
- "당신은 해적입니까?"를 반환합니다. `언어`가 `"R"`인 경우(모든 종류의 대문자)
- 그렇지 않으면 "`언어`가 무엇입니까?"를 반환합니다. `언어`가 다른 것이라면.

`````python
language = "python"  # noqa: F811
if language.lower() == "python":
    print("I love snakes!")
elif language.lower() == "r":
    print("Are you a pirate?")
else:
    print(f"What is {language}?")
`````
