## 1. 단위 테스트
<hr>

지난 장에서 우리는 Python 함수에 대해 논의했습니다. 하지만 우리의 함수가 우리가 기대하는 대로 정확히 수행되고 있는지 어떻게 확신할 수 있습니까? **단위 테스트**는 우리가 기대하는 결과를 제공하는지 확인하기 위해 함수를 테스트하는 프로세스입니다. 여기서는 개념을 간략하게 소개하겠습니다.

### `assert` 문

'assert' 문은 함수를 테스트하는 가장 일반적인 방법입니다. 테스트된 조건이 `False`인 경우 프로그램이 실패하게 됩니다. 구문은 다음과 같습니다.

``파이썬
Assert 표현식, "표현식이 False이거나 오류를 발생시키는 경우 오류 메시지."
````

`````python
assert 1 == 2, "1 is not equal to 2."  # noqa: F811
`````

두 숫자가 대략 같다고 주장하는 것도 도움이 될 수 있습니다. 컴퓨터의 부동 소수점 연산의 한계로 인해 우리가 같을 것으로 기대하는 숫자가 때때로 그렇지 않을 수 있습니다.

`````python
assert 0.1 + 0.2 == 0.3, "Not equal!"  # noqa: F811
`````

`````python
import math  # we'll learn about importing modules next chapter

assert math.isclose(0.1 + 0.2, 0.3), "Not equal!"  # noqa: F811
`````

불리언(Boolean)로 평가되는 모든 명령문을 테스트할 수 있습니다.

`````python
assert "varada" in ["mike", "tom", "tiffany"], "Instructor not present!"  # noqa: F811
`````

### 테스트 주도 개발

TDD(테스트 기반 개발)는 실제 코드를 구현하기 전에 테스트를 작성하는 곳입니다([자세한 내용은 여기](https://py-pkgs.org/chapters/04-testing#when-to-write-your-tests)). 이는 다소 반직관적으로 보일 수 있지만 실제 함수 이전에 함수에 대한 기대치를 생성하는 것입니다. 이는 여러 가지 이유로 도움이 될 수 있습니다.
- 어떤 코드를 작성해야 하는지 정확히 이해하게 될 것입니다.
- 미리 테스트를 작성해야 합니다.
- 시간이 많이 걸리는 버그가 발생하지 않습니다. 그리고,
- 작고 점진적인 코드 개선 및 추가에 중점을 두어 작업 흐름을 관리 가능하게 유지하는 데 도움이 됩니다.

일반적으로 접근 방식은 다음과 같습니다.
1. 스텁(stub) 작성: 모든 입력 매개변수를 받아들이고 올바른 데이터 타입을 반환하는 함수만 작성합니다.
2. 설계 사양을 충족하는 테스트를 작성합니다.
3. 의사 코드를 사용하여 프로그램 개요를 작성합니다.
4. 코드를 작성하고 자주 테스트하십시오.
5. 문서를 작성하세요.

### EAFP 대 LBYL

테스트 및 함수 설계와 다소 관련이 있는 철학은 EAFP 및 LBYL입니다. EAFP = "허락을 구하는 것보다 용서를 구하는 것이 더 쉽습니다(EAFP)". 코딩 용어에서: 뭔가를 try해 보고, 작동하지 않으면 오류를 잡아보세요. LBYL = "도약하기 전에 살펴보세요". 코딩 용어: 무언가를 try하기 전에 먼저 할 수 있는지 확인하세요. 이 두 약어는 코드 작성 방법에 대한 코딩 철학을 나타냅니다. 예를 살펴보겠습니다:

`````python
d = {  # noqa: F811
    "name": "Doctor Python",
    "superpower": "programming",
    "weakness": "mountain dew",
    "enemies": 10,
}
`````

`````python
# EAFP
try:
    d["address"]
except KeyError:
    print("Please forgive me!")
`````

`````python
# LBYL
if "address" in d.keys():
    d["address"]
else:
    print("Saved you before you leapt!")
`````

EAFP는 종종 Python에서 보증되지만 코딩에는 옳고 그른 방법이 없으며 상황에 따라 달라지는 경우가 많습니다. 저는 개인적으로 상황에 따라 두 가지 방식을 혼합하여 사용합니다.

## 2. 디버깅
<hr>

그렇다면 Python 코드가 작동하지 않으면 어떻게 해야 할까요? 현재 대부분의 사용자는 "수동 테스트" 또는 "탐색 테스트"를 수행할 것입니다. 코드가 정상적으로 동작할 때까지 수동으로 수정하거나, 문제를 격리하기 위해 인쇄 문을 주변에 추가할 수도 있습니다. 예를 들어, COS 126, [조건문 및 루프](http://www.cs.princeton.edu/courses/archive/fall10/cos126/locationments/loops.html)의 허가를 받아 조정된 아래 `random_walker` 코드를 고려해 보세요.

`````python
from random import random  # noqa: F405, F811, F821


def random_walker(T):  # noqa: F811
    """
    Simulates T steps of a 2D random walk, and prints the result after each step.  # noqa: F405, F811, F821
    Returns the squared distance from the origin.

    Parameters
    ----------
    T : int
        The number of steps to take.

    Returns
    -------
    float
        The squared distance from the origin to the endpoint, rounded to 2 decimals.

    Examples
    --------
    >>> random_walker(3)
    (0, -1)
    (0, 0)
    (0, -1)
    1.0
    """

    x = 0  # noqa: F811
    y = 0  # noqa: F811

    for i in range(T):
        rand = random()  # noqa: F405, F811, F821
        if rand < 0.25:
            x += 1
        if rand < 0.5:
            x -= 1
        if rand < 0.75:
            y += 1
        else:
            y -= 1
        print((x, y))
    return round((x**2 + y**2) ** 0.5, 2)


random_walker(5)
`````

위의 코드를 다시 실행하면 랜덤 워커가 절대 오른쪽으로 가지 않습니다(x 좌표는 절대 +ve가 아닙니다). 무슨 일이 일어나고 있는지 확인하기 위해 여기에 몇 가지 인쇄 문을 추가하려고 할 수 있습니다.

`````python
def random_walker(T):  # noqa: F811
    """
    Simulates T steps of a 2D random walk, and prints the result after each step.  # noqa: F405, F811, F821
    Returns the squared distance from the origin.

    Parameters
    ----------
    T : int
        The number of steps to take.

    Returns
    -------
    float
        The squared distance from the origin to the endpoint, rounded to 2 decimals.

    Examples
    --------
    >>> random_walker(3)
    (0, -1)
    (0, 0)
    (0, -1)
    1.0
    """

    x = 0  # noqa: F811
    y = 0  # noqa: F811

    for i in range(T):
        rand = random()  # noqa: F405, F811, F821
        print(rand)
        if rand < 0.25:
            print("I'm going right!")
            x += 1
        if rand < 0.5:
            print("I'm going left!")
            x -= 1
        if rand < 0.75:
            y += 1
        else:
            y -= 1
        print((x, y))
    return round((x**2 + y**2) ** 0.5, 2)


random_walker(5)
`````

아! 우리는 ``나는 오른쪽으로 가고 있어요!'' 이후에도 즉시 ``나는 왼쪽으로 가고 있어요!''라는 메시지를 받는 것을 봅니다. 문제는 `if` 문에 있습니다. 초기 `if` 다음에 각 문에 `elif`를 사용해야 합니다. 그렇지 않으면 매번 여러 조건이 충족될 수 있습니다.

이것은 매우 간단한 디버깅 사례였으며, print 문을 추가하는 것이 항상 도움이 되거나 효율적인 것은 아닙니다. 대안으로 `pdb` 모듈을 사용할 수 있습니다. [`pdb`는 Python 디버거입니다](https://docs.python.org/3/library/pdb.html). 표준 라이브러리에 포함되어 있습니다. `breakpoint()`를 사용하여 `pdb`를 활용하고 코드의 어느 지점에나 "중단점"을 설정한 다음 변수를 검사할 수 있습니다. 디버거 콘솔과 상호 작용하는 데 도움이 필요하면 `pdb` 문서 [여기](https://docs.python.org/3/library/pdb.html) 및 이 [치트시트](https://appletree.or.kr/quick_reference_cards/Python/Python%20Debugger%20Cheatsheet.pdf)를 참조하세요.

브리티시 컬럼비아 대학의 데이터 과학 석사 프로그램(MDS)에서 학생과 강사에 대한 정보를 저장하기 시작한다고 가정해 보겠습니다.

또한 우리는 회원의 이름과 성에서 전체 이름을 추출할 수 있기를 원하지만 이 정보를 다시 기록할 필요는 없습니다. 다음과 같은 함수이 유용할 수 있습니다.

### 클래스 만들기

위의 내용은 매우 비효율적이었습니다. 우리가 원하는 객체가 많아지고 객체가 더 복잡해질수록(더 많은 데이터, 더 많은 함수) 이 문제는 더욱 악화된다는 것을 상상할 수 있습니다! 그러나 이것은 수업에 대한 완벽한 사용 사례입니다! 클래스는 객체(이 경우 MDS 멤버)를 생성하기 위한 **청사진**으로 생각할 수 있습니다.

**용어 경고**:
- 클래스 데이터 = "속성"
- 클래스 함수 = "메서드"

**구문 경고**:
- `class` 키워드와 이름, 콜론(`:`)을 사용하여 클래스를 정의합니다.

`````python
class mds_member:  # noqa: F811
    pass
`````

`````python
mds_1 = mds_member()  # noqa: F811
type(mds_1)
`````

예를 들어 인스턴스에 데이터를 추가하기 위해 새 인스턴스를 만들 때마다 실행되는 `__init__` 메서드를 클래스에 추가할 수 있습니다. `mds_member` 클래스에 `__init__` 메소드를 추가해 보겠습니다. `self`는 클래스의 인스턴스를 참조하며 항상 클래스 메서드의 첫 번째 인수로 전달되어야 합니다.

`````python
class mds_member:  # noqa: F811
    def __init__(self, first, last):  # noqa: F811
        # the below are called "attributes"
        self.first = first
        self.last = last
        self.email = first.lower() + "." + last.lower() + "@mds.com"
`````

`````python
mds_1 = mds_member("Varada", "Kolhatkar")  # noqa: F811
print(mds_1.first)
print(mds_1.last)
print(mds_1.email)
`````

전체 이름을 얻으려면 이전에 정의한 함수를 사용할 수 있습니다.

`````python
full_name(mds_1.first, mds_1.last)
`````

하지만 이를 수행하는 더 좋은 방법은 이 함수를 '메서드'로 클래스에 통합하는 것입니다.

`````python
class mds_member:  # noqa: F811
    def __init__(self, first, last):  # noqa: F811
        self.first = first
        self.last = last
        self.email = first.lower() + "." + last.lower() + "@mds.com"

    def full_name(self):  # noqa: F811
        return f"{self.first} {self.last}"
`````

### 메서드, 클래스 메서드 및 정적 메서드

지금까지 본 `메서드`는 때때로 "일반" `메소드`라고 불리며 클래스의 인스턴스에 대해 작동합니다(즉, `self`를 인수로 사용). 실제 클래스에 작용하는 '클래스 메소드'도 있습니다. `클래스 메소드`는 종종 "대체 생성자"로 사용됩니다. 예를 들어, 일반적으로 누군가가 다음과 같이 쉼표로 구분된 이름으로 우리 클래스를 사용하기를 원한다고 가정해 보겠습니다.

`````python
name = "Tom,Beuzen"  # noqa: F811
`````

안타깝게도 해당 사용자는 다음 작업을 수행할 수 없습니다.

`````python
mds_member(name)
`````

우리 클래스를 사용하려면 이 문자열을 `first`와 `last`로 구문 분석해야 합니다.

`````python
first, last = name.split(",")
print(first)
print(last)
`````

그런 다음 우리 클래스의 인스턴스를 만들 수 있습니다.

`````python
mds_1 = mds_member(first, last)  # noqa: F811
`````

이것이 우리 코드 사용자의 일반적인 사용 사례라면 우리는 그들이 우리 클래스를 사용하기 전에 매번 데이터를 강제할 필요가 없기를 바랍니다. 대신 `클래스 메소드`를 사용하여 사용 사례를 용이하게 할 수 있습니다. `클래스 메소드`를 사용하려면 다음 두 가지 작업을 수행해야 합니다.
1. 데코레이터 `@classmethod`를 사용하여 메소드를 `클래스 메소드`로 식별합니다(데코레이터에 대해 조금 더 자세히 설명).
2. 첫 번째 인수로 `self` 대신 `cls`를 전달합니다.

`````python
class mds_member:  # noqa: F811
    role = "MDS member"  # noqa: F811
    campus = "UBC"  # noqa: F811

    def __init__(self, first, last):  # noqa: F811
        self.first = first
        self.last = last
        self.email = first.lower() + "." + last.lower() + "@mds.com"

    def full_name(self):  # noqa: F811
        return f"{self.first} {self.last}"

    @classmethod
    def from_csv(cls, csv_name):  # noqa: F405, F811, F821
        first, last = csv_name.split(",")  # noqa: F405, F811, F821
        return cls(first, last)
`````

이제 쉼표로 구분된 값을 직접 사용할 수 있습니다!

`````python
mds_1 = mds_member.from_csv("Tom,Beuzen")  # noqa: F811
mds_1.full_name()
`````

'정적 메소드'라는 세 번째 종류의 메소드가 있습니다. `정적 메소드`는 인스턴스나 클래스에서 작동하지 않으며 단순한 함수일 뿐입니다. 하지만 그것들이 우리 수업과 어느 정도 관련이 있기 때문에 우리 수업에 포함시키고 싶을 수도 있습니다. `@staticmethod` 데코레이터를 사용하여 정의됩니다.

`````python
class mds_member:  # noqa: F811
    role = "MDS member"  # noqa: F811
    campus = "UBC"  # noqa: F811

    def __init__(self, first, last):  # noqa: F811
        self.first = first
        self.last = last
        self.email = first.lower() + "." + last.lower() + "@mds.com"

    def full_name(self):  # noqa: F811
        return f"{self.first} {self.last}"

    @classmethod
    def from_csv(cls, csv_name):  # noqa: F405, F811, F821
        first, last = csv_name.split(",")  # noqa: F405, F811, F821
        return cls(first, last)

    @staticmethod
    def is_quizweek(week):  # noqa: F811
        return True if week in [3, 5] else False
`````

`is_quizweek()` 메소드는 `self` 인수를 허용하거나 사용하지 않습니다. 하지만 여전히 MDS와 관련되어 있으므로 여기에 포함시키고 싶을 수도 있습니다.

`````python
mds_1 = mds_member.from_csv("Tom,Beuzen")  # noqa: F811
print(f"Is week 1 a quiz week? {mds_1.is_quizweek(1)}")
print(f"Is week 3 a quiz week? {mds_1.is_quizweek(3)}")
`````

### 데코레이터

데코레이터는 꽤 복잡한 주제일 수 있습니다. 데코레이터에 대한 자세한 내용은 [여기](https://realpython.com/primer-on-python- decorators/)에서 확인할 수 있습니다. 간단히 말하면, 함수/메서드를 추가 함수으로 "장식"하는 것입니다. 데코레이터는 다른 함수을 취하고 함수을 추가하는 함수로 생각할 수 있습니다.

예를 들어 데코레이터를 만들어 보겠습니다. 함수는 Python의 데이터 타입이며 다른 함수에 전달될 수 있다는 점을 기억하세요. 따라서 데코레이터는 단순히 함수를 인수로 사용하고, 여기에 더 많은 함수을 추가하고, 실행될 수 있는 "장식된 함수"를 반환합니다.

`````python
# some function we wish to decorate
def original_func():  # noqa: F811
    print("I'm the original function!")


# a decorator
def my_decorator(original_func):  # takes our original function as input  # noqa: F811

    def wrapper():  # wraps our original function with some extra functionality  # noqa: F811
        print(f"A decoration before {original_func.__name__}.")
        result = original_func()  # noqa: F811
        print(f"A decoration after {original_func.__name__}.")
        return result

    return (
        wrapper  # returns the unexecuted wrapper function which we can can excute later
    )
`````

`my_ decorator()` 함수는 원래 함수의 데코레이팅된 버전인 함수를 반환합니다.

`````python
my_decorator(original_func)
`````

함수가 반환되었으므로 괄호를 추가하여 실행할 수 있습니다.

`````python
my_decorator(original_func)()
`````

데코레이터를 사용하여 임의의 함수을 꾸밀 수 있습니다.

`````python
def another_func():  # noqa: F811
    print("I'm a different function!")


my_decorator(another_func)()
`````

데코레이터를 호출하는 구문은 읽기 쉽지 않습니다. 대신 `@` 기호를 "구문적 설탕"으로 사용하여 데코레이터의 가독성과 재사용성을 향상시킬 수 있습니다.

`````python
@my_decorator
def one_more_func():  # noqa: F811
    print("One more function...")


one_more_func()
`````

좋아요, 좀 더 유용한 것을 만들어 보겠습니다. 임의 함수의 실행 시간을 측정하는 데코레이터를 만듭니다.

`````python
import time  # import the time module, we'll learn about imports next chapter


def timer(my_function):  # the decorator  # noqa: F811

    def wrapper():  # the added functionality  # noqa: F811
        t1 = time.time()  # noqa: F811
        result = my_function()  # the original function  # noqa: F811
        t2 = time.time()  # noqa: F811
        print(
            f"{my_function.__name__} ran in {t2 - t1:.3f} sec"
        )  # print the execution time
        return result

    return wrapper
`````

`````python
@timer
def silly_function():  # noqa: F811
    for i in range(10_000_000):
        if (i % 1_000_000) == 0:
            print(i)
        else:
            pass


silly_function()
`````

`classmethod` 및 `staticmethod`와 같은 Python의 내장 데코레이터는 C로 코딩되어 있으므로 여기에 표시하지 않습니다. 저는 자체 데코레이터를 자주 만들지는 않지만 항상 내장된 데코레이터를 사용합니다.

이제 `grade`라는 클래스에 `instance 속성`을 추가해 보겠습니다. 다음과 같이 하고 싶은 유혹을 느낄 수도 있습니다.

`````python
class mds_student(mds_member):  # noqa: F811
    role = "MDS student"  # noqa: F811

    def __init__(self, first, last, grade):  # noqa: F811
        self.first = first
        self.last = last
        self.email = first.lower() + "." + last.lower() + "@mds.com"
        self.grade = grade


student_1 = mds_student("John", "Smith", "B+")  # noqa: F811
print(student_1.email)
print(student_1.grade)
`````

괄호는 생략될 수 있으며(종종 생략됨), '튜플'은 쉼표를 사용하여 정의된 대로 암시적으로 반환됩니다.

`````python
class mds_student(mds_member):  # noqa: F811
    role = "MDS student"  # noqa: F811

    def __init__(self, first, last, grade):  # noqa: F811
        super().__init__(first, last)
        self.grade = grade


student_1 = mds_student("John", "Smith", "B+")  # noqa: F811
print(student_1.email)
print(student_1.grade)
`````

놀라운! 상속이 얼마나 강력한지 알 수 있기를 바랍니다. `add_course()`와 `remove_course()`라는 두 개의 새로운 메소드가 있는 `mds_instructor`라는 또 다른 하위 클래스를 만들어 보겠습니다.

`````python
class mds_instructor(mds_member):  # noqa: F811
    role = "MDS instructor"  # noqa: F811

    def __init__(self, first, last, courses=None):  # noqa: F811
        super().__init__(first, last)
        self.courses = [] if courses is None else courses

    def add_course(self, course):  # noqa: F811
        self.courses.append(course)

    def remove_course(self, course):  # noqa: F811
        self.courses.remove(course)
`````

`````python
instructor_1 = mds_instructor("Tom", "Beuzen", ["511", "561", "513"])  # noqa: F811
print(instructor_1.full_name())
print(instructor_1.courses)
`````

`````python
instructor_1.add_course("591")
instructor_1.remove_course("513")
instructor_1.courses
`````

### 게터/세터/삭제자

Python 클래스에 관해 이야기할 가져오기 주제가 하나 더 있는데, 그것은 getters/setters/deleters입니다. 이러한 조치의 필요성은 사례를 통해 가장 잘 설명됩니다. 다음은 이전 `mds_member` 클래스의 간단한 버전입니다:

`````python
class mds_member:  # noqa: F811
    def __init__(self, first, last):  # noqa: F811
        self.first = first
        self.last = last
        self.email = first.lower() + "." + last.lower() + "@mds.com"

    def full_name(self):  # noqa: F811
        return f"{self.first} {self.last}"
`````

`````python
mds_1 = mds_member("Tom", "Beuzen")  # noqa: F811
print(mds_1.first)
print(mds_1.last)
print(mds_1.email)
print(mds_1.full_name())
`````

내가 이 클래스 인스턴스의 이름을 잘못 입력하여 수정하고 싶었다고 가정해 보겠습니다. 무슨 일이 일어나는지 지켜보세요...

`````python
mds_1.first = "Tomas"
print(mds_1.first)
print(mds_1.last)
print(mds_1.email)
print(mds_1.full_name())
`````

아아... 이메일이 새 이름으로 업데이트되지 않았습니다! `full_name()` 메서드에서는 현재 `이름`과 `성` 이름만 호출하기 때문에 이 문제가 발생하지 않았습니다. 여기서 가장 좋은 일은 `full_name()`에 대해 했던 것처럼 `email()`에 대한 메소드를 만드는 것이라고 생각할 수도 있습니다. 그러나 이는 여러 가지 이유로 잘못된 코딩입니다. 예를 들어 코드 사용자는 `email` 속성에 대한 모든 호출을 `email()` 메서드에 대한 호출로 변경해야 함을 의미합니다. 우리는 이를 소프트웨어에 대한 획기적인 변경이라고 부르며 가능한 경우 이를 피하고 싶습니다. 대신에 우리가 할 수 있는 일은 `email`을 메소드처럼 정의하고 `@property` 데코레이터를 사용하여 속성으로 유지하는 것입니다.

`````python
class mds_member:  # noqa: F811
    def __init__(self, first, last):  # noqa: F811
        self.first = first
        self.last = last

    def full_name(self):  # noqa: F811
        return f"{self.first} {self.last}"

    @property
    def email(self):  # noqa: F811
        return self.first.lower() + "." + self.last.lower() + "@mds.com"
`````

`````python
mds_1 = mds_member("Tom", "Beuzen")  # noqa: F811
mds_1.first = "Tomas"
print(mds_1.first)
print(mds_1.last)
print(mds_1.email)
print(mds_1.full_name())
`````

원한다면 `full_name()` 메소드를 사용해 동일한 작업을 수행할 수도 있습니다...

`````python
class mds_member:  # noqa: F811
    def __init__(self, first, last):  # noqa: F811
        self.first = first
        self.last = last

    @property
    def full_name(self):  # noqa: F811
        return f"{self.first} {self.last}"

    @property
    def email(self):  # noqa: F811
        return self.first.lower() + "." + self.last.lower() + "@mds.com"
`````

하지만 지금 전체 이름을 변경하고 싶다면 어떻게 될까요?

오류가 발생합니다... 클래스 인스턴스는 전달된 값으로 무엇을 해야할지 모릅니다. 이상적으로는 클래스 인스턴스가 이 전체 이름 정보를 사용하여 `self.first` 및 `self.last`를 업데이트하는 것이 좋습니다. 이 작업을 처리하려면 데코레이터 `@<attribute>.setter`를 사용하여 정의된 `setter`가 필요합니다.

`````python
class mds_member:  # noqa: F811
    def __init__(self, first, last):  # noqa: F811
        self.first = first
        self.last = last

    @property
    def full_name(self):  # noqa: F811
        return f"{self.first} {self.last}"

    @full_name.setter
    def full_name(self, name):  # noqa: F811
        first, last = name.split(" ")
        self.first = first
        self.last = last

    @property
    def email(self):  # noqa: F811
        return self.first.lower() + "." + self.last.lower() + "@mds.com"
`````

`````python
mds_1 = mds_member("Tom", "Beuzen")  # noqa: F811
mds_1.full_name = "Thomas Beuzen"
print(mds_1.first)
print(mds_1.last)
print(mds_1.email)
print(mds_1.full_name)
`````

거의 다 왔어요! 정보를 얻고 정보를 설정하는 것에 대해 이야기했는데 정보를 삭제하는 것은 어떻습니까? 이는 일반적으로 일부 정리를 수행하는 데 사용되며 `@<attribute>.deleter` 데코레이터로 정의됩니다. 나는 이 방법을 거의 사용하지 않지만 여러분이 이 방법을 보셨으면 합니다.

`````python
class mds_member:  # noqa: F811
    def __init__(self, first, last):  # noqa: F811
        self.first = first
        self.last = last

    @property
    def full_name(self):  # noqa: F811
        return f"{self.first} {self.last}"

    @full_name.setter
    def full_name(self, name):  # noqa: F811
        first, last = name.split(" ")
        self.first = first
        self.last = last

    @full_name.deleter
    def full_name(self):  # noqa: F811
        print("Name deleted!")
        self.first = None
        self.last = None

    @property
    def email(self):  # noqa: F811
        return self.first.lower() + "." + self.last.lower() + "@mds.com"
`````

`````python
mds_1 = mds_member("Tom", "Beuzen")  # noqa: F811
delattr(mds_1, "full_name")
print(mds_1.first)
print(mds_1.last)
`````
