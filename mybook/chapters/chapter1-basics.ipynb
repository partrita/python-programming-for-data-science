### 산술 연산자

다음은 Python의 일반적인 산술 연산에 대한 구문 표입니다.

| 운영자 |   설명 |
| :------: | :---------------: |

|   `-` |   빼기 |
|   `*` |  곱셈 |
|   `/` |     부문 |
|   `**` |  지수 |

이러한 연산자를 숫자 유형에 적용하고 결과를 관찰해 보겠습니다.

`````python
1 + 2 + 3 + 4 + 5  # add
`````

`````python
2 * 3.14159  # multiply
`````

`````python
2**10  # exponent
`````

나누기는 예상과 다른 `dtype`을 생성할 수 있으며 `int`를 `float`로 변경합니다.

`````python
int_2 = 2  # noqa: F811
type(int_2)
`````

`````python
int_2 / int_2  # divison
`````

`````python
type(int_2 / int_2)
`````

그러나 `//` 구문을 사용하면 "정수 나누기"(일명 "정수 나눗셈(Floor Division)")를 수행하고 `int` 데이터 타입을 유지할 수 있으며 항상 내림됩니다.

`````python
101 / 2
`````

`````python
101 // 2  # "floor division" - always rounds down
`````

우리는 이것을 "정수 나누기" 또는 "정수 나눗셈(Floor Division)"라고 부릅니다. 나눗셈 결과에 대해 `int`를 호출하여 가장 가까운 정수로 내림하거나 결과를 "바닥"으로 만드는 것과 같기 때문입니다.

`````python
int(101 / 2)
`````

`%` "나머지(Modulo)" 연산자는 나눗셈 후 나머지를 제공합니다.

`````python
100 % 2  # "100 mod 2", or the remainder when 100 is divided by 2
`````

`````python
101 % 2  # "101 mod 2", or the remainder when 101 is divided by 2
`````

`````python
100.5 % 2
`````

### 시퀀스 인덱싱과 슬라이싱

대괄호 구문을 사용하여 리스트, 튜플 또는 문자열 내부의 값에 액세스할 수 있습니다. Python은 *0부터 시작하는 인덱싱*을 사용합니다. 이는 리스트의 첫 번째 요소가 위치 1이 아닌 위치 0에 있음을 의미합니다.

`````python
my_list
`````

`````python
my_list[0]
`````

`````python
my_list[2]
`````

`````python
len(my_list)
`````

`````python
my_list[5]
`````

리스트의 끝에서부터 거꾸로 계산하려면 음수 인덱스를 사용할 수 있습니다.

`````python
my_list
`````

`````python
my_list[-1]
`````

`````python
my_list[-2]
`````

콜론`:`을 사용하여 하위 시퀀스에 액세스할 수 있습니다. 이것을 "슬라이싱"이라고 합니다.

## 7. 조건문
<hr>

[조건문](https://docs.python.org/3/tutorial/controlflow.html)을 사용하면 프로그램 상태에 따라 특정 코드 블록만 실행되는 프로그램을 작성할 수 있습니다. 몇 가지 예를 살펴보고 키워드, 구문 및 들여쓰기를 기록해 보겠습니다.

`````python
name = "Tom"  # noqa: F811

if name.lower() == "tom":
    print("That's my name too!")
elif name.lower() == "santa":
    print("That's a funny name.")
else:
    print(f"Hello {name}! That's a cool name!")
print("Nice to meet you!")
`````

주목해야 할 주요 사항:
- `if`, `elif`, `else` 키워드를 사용하세요.
- 콜론`:`은 각 조건식을 끝냅니다.
- 들여쓰기(4개의 빈 공간)는 코드 블록을 정의합니다.
- 'if' 문에서 조건문이 'True'를 반환하는 첫 번째 블록이 실행되고 프로그램이 'if' 블록을 종료합니다.
- `if` 문에는 반드시 `elif`나 `else`가 필요하지 않습니다.
- `elif`를 사용하면 여러 조건을 확인할 수 있습니다.
- `else`를 사용하면 다른 모든 조건이 `False`인 경우 기본 블록을 평가할 수 있습니다.
- 전체 `if` 문의 끝은 첫 번째 `if` 키워드와 동일한 수준으로 들여쓰기가 반환되는 곳입니다.

If 문은 서로 **중첩**될 수도 있습니다.

`````python
name = "Super Tom"  # noqa: F811

if name.lower() == "tom":
    print("That's my name too!")
elif name.lower() == "santa":
    print("That's a funny name.")
else:
    print(f"Hello {name}! That's a cool name.")
    if name.lower().startswith("super"):
        print("Do you really have superpowers?")

print("Nice to meet you!")
`````

### 불리언 테스트(Truth Value Testing)

Python에서는 `if` 및 `while`(다음 장) 문에 사용하기 위해 모든 객체의 "진실"을 테스트할 수 있습니다.
- `True` 값: `False` 값을 갖는 `bool` 객체이거나 `len()` == 0을 갖지 않는 한 모든 객체는 `True`를 반환합니다.
- `False` 값: `None`, `False`, `0`, 빈 시퀀스 및 컬렉션: `''`, `()`, `[]`, `{}`, `set()`

::: {.callout-tip}
[여기 문서](https://docs.python.org/3/library/stdtypes.html#truth-value-testing)에서 자세히 알아보세요.
:::

`````python
x = 1  # noqa: F811

if x:
    print("I'm truthy!")
else:
    print("I'm falsey!")
`````

`````python
x = False  # noqa: F811

if x:
    print("I'm truthy!")
else:
    print("I'm falsey!")
`````

`````python
x = []  # noqa: F811

if x:
    print("I'm truthy!")
else:
    print("I'm falsey!")
`````

|표현식|결과|세부정보|
|---|---|---|
|A 또는 B|A가 'True'이면 A else B|A가 'False'인 경우에만 B가 실행됩니다.|
|A 및 B|A가 'False'이면 A else B|A가 'True'인 경우에만 B가 실행됩니다.|
